```python
import numpy as np
import matplotlib.pyplot as plt
import ast
import os
```

```python
def tl_loader(path_to_file):
    list_to_return = []
    with open(path_to_file, 'r') as file:
        data_text = file.read()
    load_list = ast.literal_eval(data_text)
    for element in load_list:
        list_to_return.append(element[1])
    return list_to_return

def thr_main_loader(path_to_file):
    with open(path_to_file, 'r') as file:
        data_text = file.read()
    data_text = data_text.replace("[", "").replace("]", "")
    sections = data_text.split('\n')
    list_to_return = []
    for section in sections:
        values = section.split(',')
        list_to_return.extend([float(item.strip()) for item in values if item.strip()])
    return list_to_return

def power_loader(path_to_file):
   with open(path_to_file, 'r') as file:
      data_text = file.read()
   data_text = data_text.replace("[", "").replace("]", "")
   sections = data_text.split('\n')
   array_to_return = []
   for section in sections:
      values = section.split(',')
      section_data = [float(item.strip()) for item in values if item.strip()]
      if section_data:
         array_to_return.append(section_data)
   return np.array(array_to_return)
```

```python
power_1_opt = power_loader(os.getcwd() +'/result/sf1_opt.csv')
power_3_opt = power_loader(os.getcwd() +'/result/sf3_opt.csv')
power_5_opt = power_loader(os.getcwd() +'/result/sf5_opt.csv')
power_8_opt = power_loader(os.getcwd() +'/result/sf8_opt.csv')
power_10_opt = power_loader(os.getcwd() +'/result/sf10_opt.csv')
```

```python
power_1_avg_o = np.mean(power_1_opt, axis=0);print(power_1_avg_o.sum())
power_3_avg_o = np.mean(power_3_opt, axis=0);print(power_3_avg_o.sum())
power_5_avg_o = np.mean(power_5_opt, axis=0);print(power_5_avg_o.sum())
power_8_avg_o = np.mean(power_8_opt, axis=0);print(power_8_avg_o.sum())
power_10_avg_o = np.mean(power_10_opt, axis=0);print(power_10_avg_o.sum())
```

```python
plt.plot(np.arange(1,len(power_1_avg_o)+1), power_1_avg_o, color='#0a9396', label='SF=1')
plt.plot(np.arange(1,len(power_3_avg_o)+1), power_3_avg_o, color='#ee9b00', label='SF=3')
plt.plot(np.arange(1,len(power_5_avg_o)+1), power_5_avg_o, color='#ca6702', label='SF=5')
plt.plot(np.arange(1,len(power_8_avg_o)+1), power_8_avg_o, color='#ae2012', label='SF=8')
plt.plot(np.arange(1,len(power_10_avg_o)+1), power_10_avg_o, color='#001219', label='SF=10')
plt.legend()
plt.xlabel('Query number')
plt.ylabel('Execution time (seconds)')
plt.title("Optimized queries")
plt.show()
```
